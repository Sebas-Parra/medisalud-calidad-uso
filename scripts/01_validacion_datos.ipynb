# Escenario 7 - Paso 3: Validacion de datos

Valida `data/logs_hce.csv` y `data/encuesta_satisfaccion.csv` antes de calcular cualquier metrica: nulos, rangos logicos, duplicados y resumen descriptivo.

```python
import pandas as pd

df = pd.read_csv("../data/logs_hce.csv")
df.head()
```

## 1. Valores nulos

```python
print("Valores nulos por columna:")
print(df.isnull().sum())
```

## 2. Rangos logicos

Un registro de nota de evolucion no deberia tomar menos de 0s ni mas de 120s (un valor asi indicaria una sesion abandonada o un error de instrumentacion, no un tiempo de tarea real).

```python
print("Tiempos fuera de rango (negativos o mayores a 120s):")
print(df[(df["tiempo_segundos"] < 0) | (df["tiempo_segundos"] > 120)])
```

## 3. Duplicados

```python
print("Eventos duplicados:", df.duplicated(subset=["evento_id"]).sum())
```

## 4. Resumen descriptivo

```python
print("Resumen estadistico de tiempo_segundos:")
print(df["tiempo_segundos"].describe())
```

## 5. Validacion de la encuesta de satisfaccion

Se repite la misma disciplina de validacion (nulos, rango logico, duplicados) sobre la segunda fuente de datos generada en este escenario.

```python
encuesta = pd.read_csv("../data/encuesta_satisfaccion.csv")

print("Valores nulos por columna:")
print(encuesta.isnull().sum())
print()

print("Puntajes CSAT fuera de rango (deben estar entre 1 y 5):")
print(encuesta[(encuesta["puntaje_csat"] < 1) | (encuesta["puntaje_csat"] > 5)])
print()

print("Respuestas duplicadas:", encuesta.duplicated(subset=["respuesta_id"]).sum())
print()

print("Resumen estadistico de puntaje_csat:")
print(encuesta["puntaje_csat"].describe())
```
